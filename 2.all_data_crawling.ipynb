```python
import pickle
with open("scraper_link.pickle", 'rb') as f:
     link = pickle.load(f)
```

```python
# link 잘 불러왔는지 검증
lengths = [len(value) for value in link.values()]
print(f"link 수 : {lengths}", f"| 국가 수 : {len(lengths)}")
```

```python
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm 
# 최종 결과를 저장할 딕셔너리
result_data = {}

# 각 언어에 대해 URL을 순회하며 페이지의 내용을 추출
for lang, urls in link.items():  # tqdm을 사용하여 로딩바 표시
    lang_data = {}
    
    for url in tqdm(urls, desc=f'Processing {lang} URLs', leave=False):  # tqdm을 사용하여 로딩바 표시
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')
        
        # <p> 태그 전체를 찾아서 그 내용을 리스트에 저장
        paragraphs = [p.get_text() for p in soup.find_all('p')]
        # 언어 데이터 딕셔너리에 결과 저장
        lang_data[url] = paragraphs
    
    # 최종 결과 딕셔너리에 언어 데이터 저장
    result_data[lang] = lang_data

# 최종 결과를 JSON 형식으로 수집
import json
json_result = json.dumps(result_data, indent=2, ensure_ascii=False)
```

```python
print(json_result)
```

```python
# JSON save
with open("data.json", "w") as f:
    f.write(json_result)
```
